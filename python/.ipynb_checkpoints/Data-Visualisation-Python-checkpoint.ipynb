# Visualisation libraries in Python

```python
import pandas as pd

df = pd.read_csv("../GLB-Ts.csv")
```

## Matplotlib

tutorials: https://matplotlib.org/users/pyplot_tutorial.html

```python
import matplotlib.pyplot as plt

plt.plot(df.Year, df.MAM)
plt.xlabel('Years')
plt.ylabel('MAM Temperature')
plt.show()
```

## Plotly
to install: conda install plotly

tutorials: https://plot.ly/python/

```python
import plotly as py
from plotly.graph_objs import Scatter, Layout

py.offline.plot({
    "data": [Scatter(x = df.Year,
    y = df.MAM)],
    "layout": Layout(title="Temperature evolution over the years")
})
```

## Seaborn

tutorials: https://seaborn.pydata.org/

```python
import seaborn as sns

sns.jointplot(x="Year", y="MAM", data=df)
```

## Bokeh

tutorials: http://bokeh.pydata.org/en/latest/docs/user_guide/plotting.html

```python
from bokeh.plotting import figure, output_file, show

p = figure(plot_width=400, plot_height=400)

p.line(df.Year, df.MAM, line_width=2)

show(p)
```

## DataShader
tutorials: http://datashader.readthedocs.io/en/latest/

```python
import datashader as ds
import datashader.transfer_functions as tf

cvs = ds.Canvas(plot_width=400, plot_height=400)
agg = cvs.points(df, 'Year', 'MAM')
img = tf.shade(agg, cmap=['lightblue', 'darkblue'], how='log')
img
```

## Altair

tutorials: https://altair-viz.github.io/

```python
from altair import Chart

Chart(df).mark_circle().encode(
    x='Year',
    y='MAM',
)
```
